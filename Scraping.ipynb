```python
import os
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

# URL de la página web donde se encuentran los enlaces a los PDF
url = 'https://www.corteconstitucional.gov.co/secretaria/autoss/'

# Directorio local donde quieres guardar los PDFs
save_dir = '/Users/oscarperez/Library/CloudStorage/Dropbox/Laboral/Acobur/Autos-Corte-Constitucional/descargas_pdfs/descargas_pdfs'

# Asegúrate de que save_dir exista o créalo
if not os.path.exists(save_dir):
    os.makedirs(save_dir)

# Realizar la solicitud HTTP a la página de enlaces
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Encontrar todos los enlaces en la página
for link in soup.find_all('a', href=True):
    # Construir la URL completa para el enlace si es necesario
    href = urljoin(url, link['href'])
    if href.lower().endswith('.pdf'):
        # Realizar una solicitud GET para el archivo PDF
        pdf_response = requests.get(href, stream=True)
        
        # Extraer el nombre del archivo PDF del enlace
        pdf_name = os.path.join(save_dir, href.split('/')[-1])
        
        # Guardar el PDF en el disco local
        with open(pdf_name, 'wb') as pdf_file:
            for chunk in pdf_response.iter_content(chunk_size=1024):
                # Escribir el contenido del PDF en el archivo
                if chunk:
                    pdf_file.write(chunk)
```
